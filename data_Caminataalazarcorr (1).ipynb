Caminata al azar

Tenemos un punto que se puede mover a la derecha con probabilidad $p$ y a la izquierda con probabilidad q, tal que $p+q=1$.

```python
import random
import matplotlib.pyplot as plt
def cami(p,pas):
    pos=0
    lis=[pos]
    for i in range(pas):
        if random.random()<=p:
            pos+=1
        else:
            pos-=1
        lis.append(pos)
    return lis
p=0.5
pas=10000
lis=cami(p,pas)
plt.plot(lis,color="#800080",linewidth=0.6)
plt.axhline(0,color="purple", lw=1)
plt.xlabel("n")
plt.ylabel("Xn")
plt.title(f"Caminata aleatoria 1D con p = {p}")
plt.show()
```

Análisis y conclusiones

- ¿Este proceso es una cadena de Markov?

Si porque para calcular la posición siguiente, el programa solo toma en cuenta la posición actual y le suma o resta $1$ al azar.


- ¿Cómo podemos obtener la distribución de la caminata?

En cada paso de la caminata aleatoria el movimiento puede ser $+1$ derecha o $-1$ izquierda. Se realizan $n$ pasos y cada uno tiene probabilidad $p$ de ser $+1$ y $q = 1-p$ de ser $-1$.

Sea $x$ el número de pasos hacia la derecha. Entonces los pasos hacia la izquierda serán $n-x$.

La posición final después de $n$ pasos es:

$X_n = (+1) x + (-1) (n-x)= x + (-n+x)=x - (n-x)=x-n+x=2x - n$


El número de pasos hacia la derecha $x$ sigue una distribución binomial porque la caminata puede interpretarse como una sucesión de $n$ ensayos independientes, donde:

En cada paso solo existen dos posibles resultados: $+1$ o $-1$

La probabilidad de obtener $+1$ es constante y igual a $p$

Por lo tanto, $x \sim Bin(n,p)$

y su función de probabilidad es:

$P(x) = \binom{n}{x} p^x (1-p)^{n-x}.$

Sabemos que la posición final es: $X_n = 2x - n$

Despejando $x$:
$x = \frac{X_n + n}{2}$

Sustituyendo en la fórmula binomial, obtenemos la probabilidad de que la caminata termine en la posición $X_n$:

$P(X_n) = \binom{n}{\frac{X_n + n}{2}} p^{\frac{X_n + n}{2}} (1-p)^{\,n-\frac{X_n + n}{2}}$

Por lo tanto, asi se obtiene la distribución binomial de la caminata después de $n$ pasos.
